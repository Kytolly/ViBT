```python
import torch
from diffusers import QwenImageEditPipeline
from PIL import Image

from vibt.scheduler import ViBTScheduler
from vibt.qwen_image import qwen_bridge_gen
```

```python
base_model_id = "Qwen/Qwen-Image-Edit"
pipe = QwenImageEditPipeline.from_pretrained(
    base_model_id, torch_dtype=torch.bfloat16
).to("cuda")
```

```python
pipe.unload_lora_weights()
pipe.load_lora_weights(
    "Yuanshi/ViBT",
    weight_name="image/image_editing.safetensors",
)

pipe.scheduler = ViBTScheduler()
```

```python
source_image = Image.open("../assets/000189825.jpg").resize((1024, 1024))
prompt = "Make this image into winter."
```

```python
pipe.scheduler.set_parameters(noise_scale=1.0, shift_gamma=5.0, seed=42)

output = qwen_bridge_gen(
    pipe,
    prompt="Change the season to winter.",
    num_inference_steps=64,
    image=source_image,
).images[0]

output
```
